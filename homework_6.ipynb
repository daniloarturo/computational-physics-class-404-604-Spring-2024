The equation for the height ($z$) as a function of time ($t$) is given by:

\begin{eqnarray}
\frac{dz}{dt} = v
\end{eqnarray}

\begin{eqnarray}
\frac{dv}{dt} = -g
\end{eqnarray}

where:
- $z$ is the height
- $t$ is the time
- $v$ is the velocity
- $g$ is the acceleration due to gravity (approximately 9.81 m/s^2)

Using the Runge-Kutta method, we can numerically solve these differential equations to obtain the trajectory of the ball.

Initial conditions: $z(0) = 0$ (initial height), $v(0) = 10$ m/s (initial velocity).


```python
import numpy as np
import matplotlib.pyplot as plt

g = 9.81  # acceleration due to gravity in m/s^2
v0 = 10   # initial velocity in m/s
z0 = 0.0  #initial height

t0 = 0
tf = 2*v0/g  # total time of flight of the motion
h = 0.01     

# Initial values 
y = [z0, v0]

# Function to calculate derivatives
def f(t, y):
    f = np.zeros(2)
    f[0] = y[1]
    f[1] = -g
    return f


#Runge-Kutta method

def rk4(t, h, y, n):
    k1 = np.zeros(n)
    k2 = np.zeros(n)
    k3 = np.zeros(n)
    k4 = np.zeros(n)
    k1 = h*f(t, y)
    k2 = h*f(t+0.5*h, y+0.5*k1)
    k3 = h*f(t+0.5*h, y+0.5*k2)
    k4 = h*f(t+h, y+k3)
    y += 1./6. * (k1 + 2*k2 + 2*k3 + k4)
    return y


t_values = np.arange(t0, tf+ h, h)
z_values = []


for t in t_values:
    z_values.append(y[0])
    y = rk4(t, h, y, 2)

# Plotting the results
plt.figure()
plt.plot(t_values, z_values)
plt.xlabel('Time (s)')
plt.ylabel('Height (m)')
plt.title('Ball Thrown Upwards:Height (m) vs Time (s) ')
plt.grid(True)
plt.show()
```
