Write a program to plot
\begin{equation}
f(x)=\pi/2+2*(sin(x)+sin(3x)/3+sin(5x)/5+....sin(nx)/n)
\end{equation}
from -$\pi$ to $\pi$. Choose n=11, 21 and 31 and plot all these three curves in one figure. Please use numpy library and matplotlib.

Then commit this to your github.

```python
import matplotlib.pylab as plt
import numpy as np


def f(x, n):
    a = np.pi / 2
    for i in range(1, n + 1, 2):
        a += 2 * (np.sin(i * x) / i)
        
    return a

# x  from -pi to pi
x = np.linspace(-np.pi, np.pi, 100)

# f(x) for different values of n
y_n11 = f(x, 11)
y_n21 = f(x, 21)
y_n31 = f(x, 31)



plt.plot(x, y_n11, label='n = 11')
plt.plot(x, y_n21, label='n = 21')
plt.plot(x, y_n31, label='n = 31')

plt.title('Plot of f(x) for n = 11, 21, 31')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
