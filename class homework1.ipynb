```python
from pylab import * 

x = np.arange(0, 2.*3.14, 0.01)


y1 = x
y2 = x - 1./6. * x**3
y3 = np.sin(x)


plt.plot(x, y1, label='y = x')
plt.plot(x, y2, label='y = x - 1/6 * x**3')
plt.plot(x, y3, label='y = sin(x)')


plt.ylim(-2, 2)


plt.xlabel('x')
plt.ylabel('y')

plt.legend()


plt.show()
```
